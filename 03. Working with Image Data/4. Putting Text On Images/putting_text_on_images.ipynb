### **Step 1: Creating the PIL Image Object**

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.new(mode="RGB", size=(200, 100), color=(0, 0, 0))

plt.imshow(img)
```

### **Step 2: Create a Draw Object**

```python
from PIL import ImageDraw

draw = ImageDraw.Draw(img)
```

### **Step 3: Preparing the Text Element**

```python
text = 'Hello World!'
```

### **Step 4: Adding Text to the Image**

```python
draw.text((50, 30), text)

plt.imshow(img)
```

### **Font Color**

```python
# Exploring Built-in Color Names in PIL
from PIL import ImageColor

colors=ImageColor.colormap

print(colors)

# Hexcode representation of  colors
```

```python
# Adding Colored Text to the Image
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# Creating an Image
img = Image.new(mode="RGB", size=(200, 100), color=(0, 0, 0))

# Create a Draw Object
draw = ImageDraw.Draw(img)

# Preparing the Text Element
text = "Hello World!"

# Adding text to the Image
draw.text((50, 30), text, fill=(255, 140, 0))

# Display the image
plt.imshow(img)
```

### **Font Type and Size**

```python
from PIL import Image, ImageDraw, ImageFont

# Creating an Image
img = Image.new(mode="RGB", size=(200, 100), color=(0, 0, 0))

# Create a Draw Object
draw = ImageDraw.Draw(img)



# Preparing the Text Element
text="Hello World!"
# ** Defining Font Type and Size
font = ImageFont.truetype('Fonts/RubikDistressed-Regular.ttf', 15)

# Adding text to the Image
draw.text((50, 30), text=text, font=font, fill=(255, 140, 0))

# Display the image
plt.imshow(img)
```

## **Creating Images with Multiple Text Elements**

```python
from PIL import Image, ImageDraw, ImageFont

# Creating an Image
img = Image.new(mode="RGB", size=(400, 300), color=(0, 0, 0))

# Create a Draw Object
draw = ImageDraw.Draw(img)

# Preparing the Text Element
text_1 = 'Image Manipulation'
text_2 = 'Python'
text_3 = 'ThatAIGuy'

# ** Defining Font Type and Size
font_1 = ImageFont.truetype('Fonts/CrimsonText-Regular.ttf', 40)
font_2 = ImageFont.truetype('Fonts/Pacifico-Regular.ttf', 50)
font_3 = ImageFont.truetype('Fonts/Inconsolata-Bold.ttf', 30)

# Preparing the Text Element
draw.text((50, 30), text_1, font=font_1, fill='hotpink')
draw.text((150, 80), text_2, font=font_2, fill=(255, 140, 0))
draw.text((200, 230), text_3, font=font_3, fill='aqua')

# Save the image
img.save('images/transformed/image_manipulation.jpg')

# Display the image
plt.imshow(img)
```

## **Creating Another Memes**

```python
from PIL import Image, ImageDraw, ImageFont

# Open the image
img = Image.open("images/meme/kgf.jpg")

print("Image Size:", img.size)
print("Image Mode:", img.mode)
plt.imshow(img)
```

```python
# Create a Blank Canvas

# Define the canvas size
canvas_width = img.width
canvas_height = img.height + 300  # Extra space for text

# Initialize the Canvas
canvas = Image.new(mode="RGB", size=(canvas_width, canvas_height), color=(0, 0, 0))

print("Image Size:", canvas.size)
print("Image Mode:", canvas.mode)
plt.imshow(canvas)
```

```python
# Paste the image onto the canvas
canvas.paste(img, (0, canvas.height - img.height))

plt.imshow(canvas)
```

```python
# Initialize ImageDraw
DRAW = ImageDraw.Draw(canvas)

# Preparing the Text Element
TEXT_1 = """ME TRYING TO BECOME
A DATA SCIENTIST"""

# Choose a font and size
font = ImageFont.truetype('Fonts/Inconsolata-Bold.ttf', 100)

# Add text to the image
DRAW.text((50, 50), text=TEXT_1, fill=(255, 255, 255), font=font, align="center")

plt.imshow(canvas)
```

```python
# Preparing the Text Element
TEXT_2 = """Math, Math, Math!
I DON'T LIKE! 
I AVOID!"""

# Choose a font and size
font = ImageFont.truetype('Fonts/Inconsolata-Bold.ttf', 50)

# Add text to the image
DRAW.text((500, 500), text=TEXT_2, fill=(255, 255, 255), font=font, align="center")

plt.imshow(canvas)
```

```python
# Preparing the Text Element
TEXT_3 = """BUT DATA SCIENCE 
REQUIRE MATHS...
I CAN'T AVOID!"""

# Choose a font and size
font = ImageFont.truetype('Fonts/Inconsolata-Bold.ttf', 50)

# Add text to the image
DRAW.text((500, 1100), text=TEXT_3, fill=(255, 255, 255), font=font, align="center")

plt.imshow(canvas)
```

```python
canvas.save("images/transformed/kgf_meme.jpg")
```

### **Creating GIFs**

```python
from PIL import Image, ImageDraw, ImageFont

img_1 = Image.new(mode="RGB", size=(400, 300), color=(0, 0, 0))

draw_1 = ImageDraw.Draw(img_1)

text_1 = 'Image Manipulation'

font_1 = ImageFont.truetype('Fonts/CrimsonText-Regular.ttf', 40)

draw_1.text((50, 30), text_1, font=font_1, fill='hotpink')

plt.imshow(img_1)
```

```python
img_2 = Image.new(mode="RGB", size=(400, 300), color=(0, 0, 0))

draw_2 = ImageDraw.Draw(img_2)

text_2 = 'with Python'

font_2 = ImageFont.truetype('Fonts/Pacifico-Regular.ttf', 30)

draw_2.text((150, 80), text_2, font=font_2, fill=(255, 140, 0))

plt.imshow(img_2)
```

```python
img_3 = Image.new(mode="RGB", size=(400, 300), color=(0, 0, 0))

draw_3 = ImageDraw.Draw(img_3)

text_3 = 'by ThatAIGuy'

font_3 = ImageFont.truetype('Fonts/Inconsolata-Bold.ttf', 20)

draw_3.text((200, 230), text_3, font=font_3, fill='aqua')

plt.imshow(img_3)
```

```python
# create a list of image objects
image_list = [img_1, img_2, img_3]

# Save the first image as a GIF file
image_list[0].save(
            'images/transformed/animation.gif',
            save_all=True,
            append_images=image_list[1:], # append rest of the images
            duration=1000, # in milliseconds
            loop=0)
```

## **Miscellaneous Image Operations in PIL**

```python
from PIL import Image, ImageDraw, ImageFont

img = Image.open('images/original_imgs/quantumania_2.png').convert("RGBA")

background = Image.new("RGBA", img.size, (0,0,0,0))
mask = Image.new("RGBA", img.size, 0)
draw = ImageDraw.Draw(mask)

text = "Ant-Man \nand \nthe Wasp: \nQuantumania"
font = ImageFont.truetype('Fonts/Inconsolata-Bold.ttf', 200)
draw.text((250,30), text, fill='blue', font=font)

new_img = Image.composite(img, background, mask)

new_img.save('images/transformed/quantumania_text.png')

plt.imshow(new_img)

# Try saving this image in jpg format
```
