### **Step-by-Step Implementation**

#### **1. Data Preparation**

First, you need a user-item interaction matrix. This matrix will have users as rows, items as columns, and ratings (or interactions) as values.

```python
import numpy as np
import pandas as pd

# Sample data: user-item interaction matrix
data = {
    'user': [0, 0, 0, 1, 1, 1, 2, 2, 3, 3],
    'item': [0, 1, 2, 0, 1, 3, 1, 2, 0, 3],
    'rating': [5, 3, 2, 4, 2, 5, 4, 1, 3, 5]
}

df = pd.DataFrame(data)

df
```

```python
# Create the user-item interaction matrix
R = df.pivot(index='user', columns='item', values='rating').fillna(0).values

R
```

```python
# Create the user-item interaction matrix on given dataset

user_item_interaction_matrix = ratings_df.pivot(index='userId', columns='movieId', values='rating').fillna(0).values

user_item_interaction_matrix
```

```python
user_item_interaction_matrix.shape
```

#### **2. Normalize the Data**

Subtract the mean rating of each user to normalize the data.


```python
# Normalize by subtracting the mean rating of each user

user_ratings_mean = np.mean(user_item_interaction_matrix, axis=1)
R_demeaned = user_item_interaction_matrix - user_ratings_mean.reshape(-1, 1)
```

#### **3. Perform Singular Value Decomposition (SVD)**

Use SVD to factorize the normalized user-item matrix.


```python
from scipy.sparse.linalg import svds

# Perform SVD
U, sigma, Vt = svds(R_demeaned, k=2)  # k is the number of latent features
sigma = np.diag(sigma)
```

#### **4. Reconstruct the User-Item Matrix**

Reconstruct the user-item matrix to predict the ratings.

```python
# Reconstruct the predicted user-item matrix
predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)

predicted_ratings_df = pd.DataFrame(predicted_ratings, columns=ratings_df['movieId'].unique())
```

#### **5. Making Recommendations**

Create a function to recommend items to users based on the predicted ratings.

```python
def recommend_items(predictions, user_id, original_ratings, num_recommendations):
    # Get and sort the user's predicted ratings
    user_row_number = user_id  # User ID starts at 0
    sorted_user_predictions = predictions[user_row_number].argsort()[::-1]
    
    # Get the user's data and merge in the item information
    user_data = original_ratings[original_ratings.user == user_id]
    recommendations = pd.DataFrame(sorted_user_predictions).rename(columns={0: 'item'})
    
    # Recommend the highest predicted rating items that the user hasn't already rated
    recommendations = recommendations[~recommendations['item'].isin(user_data['item'])]
    return recommendations.head(num_recommendations)

# Get recommendations for a specific user
user_id = 1
num_recommendations = 3
recommendations = recommend_items(predicted_ratings, user_id, df, num_recommendations)
print(recommendations)
```

### **Explanation**

1. **Data Preparation**: We prepare the user-item interaction matrix from the given dataset.
2. **Normalize the Data**: We subtract the mean rating of each user to normalize the matrix.
3. **Perform SVD**: We perform Singular Value Decomposition on the normalized matrix.
4. **Reconstruct the Matrix**: We reconstruct the user-item matrix to predict the ratings.
5. **Making Recommendations**: We generate recommendations by sorting the predicted ratings and filtering out items the user has already rated.
